```python
def get_article(soup):
    article_list = soup.find_all('p', class_='css-exrw3m evys1bk0')
    skip = '[Sign up'
    text = ''
    for item in article_list:
        #print(item.text)
        if item.text.find(skip) != -1:
            continue
        text += ' ' + item.text
    
    return text.strip()
```

```python
a = get_article(soup)
```

```python
def get_ft_links(soup):
    url_prefix = 'https://www.ft.com'
    ft_article_links = []
    articles = soup.find_all('div', class_='o-teaser__heading')
    for item in articles:
        article_url = item.find('a').attrs['href']
        print(article_url)
        ft_article_links.append(url_prefix + article_url)
    
    return ft_article_links
```
